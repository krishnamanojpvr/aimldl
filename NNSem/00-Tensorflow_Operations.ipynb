```python
#indexing and slicing

x = tf.constant([[10, 20, 30],
                 [40, 50, 60]])

print(x[0][1])   # Output: 20
print(x[1, 2])   # Output: 60
print(x[-1, -2]) # Output: 50

x = tf.constant([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])

#tensor[start:stop:step]
print(x[::2])  # rows 0 and 2
print(x[:, 1:])  # all rows, columns 1 and 2

#tf.slice(input_, begin, size)
y = tf.slice(x, [0, 1], [2, 2])
print(y)
```

```python
x = tf.constant([[1, 2], [3, 4], [5, 6]])  # shape: (3, 2)

reshaped = tf.reshape(x, [2,3])           # new shape: (2, 3)

print(reshaped)

x = tf.constant([1, 2, 3])       # shape: (3,)

y = tf.expand_dims(x, axis=0)    # shape: (1, 3)
z = tf.expand_dims(x, axis=1)    # shape: (3, 1)

print(y, z)


x = tf.constant([[[5], [6], [7]]])   # shape: (1, 3, 1)

squeezed = tf.squeeze(x)            # shape: (3,)

print(squeezed)

x = tf.constant([[1, 2, 3], [4, 5, 6]])  # shape: (2, 3)

transposed = tf.transpose(x)           # shape: (3, 2)

print(transposed)


```

```python
# broadcasting
x = tf.random.normal([2,4])
w = tf.random.normal([4,3])
b = tf.random.normal([3])
y = x@w+b
print(y)
z = x@w+tf.broadcast_to(b,[2,3])
print(z)
```

```python
# operations on tensors
a = tf.constant([1,2,3,4])
b = tf.constant([5,6,7,8])
add = tf.add(a,b)
sub = tf.subtract(a,b)
mul = tf.multiply(a,b)
div = tf.divide(a,b)
log = tf.math.log([0.2,0.3])
exp = tf.math.exp([1.0,2.0])
print(add)
print(sub)
print(mul)
print(div)
print(log)
print(exp)
```

```python
out = tf.random.normal([4,10]) # Simulate output

y = tf.constant([1,2,2,0]) # Real labels
y = tf.one_hot(y,depth=10) # One-hot encoding
loss = tf.keras.losses.mse(y,out) # Calculate loss of each sample
print(loss)

y = y - out
y = tf.square(y)
y = tf.reduce_sum(y,axis=1)/10
print(y)
loss = tf.reduce_mean(loss) # Calculate mean loss
loss
```

```python
# tensor comparision
x = tf.random.normal([100,10])
x = tf.nn.softmax(x,axis=1)
pred = tf.argmax(x,axis=1)
print(pred)
y = tf.random.uniform([100],dtype=tf.int64,maxval=10)
print(y)
z = tf.equal(pred,y)
print(z)
# find accuracy
d = tf.cast(z,dtype=tf.int64)
print(tf.reduce_sum(d))
```

```python
# meshgrid
import matplotlib.pyplot as plt
x = tf.constant([1, 2, 3])
y = tf.constant([10, 20])
X, Y = tf.meshgrid(x, y)
print("X:\n", X.numpy())
print("Y:\n", Y.numpy())
plt.scatter(X, Y)
plt.title("Points created by meshgrid")
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Generate 1D coordinates
x = tf.linspace(-8., 8., 100)
y = tf.linspace(-8., 8., 100)
x, y = tf.meshgrid(x, y)
z = tf.sqrt(x**2 + y**2)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x.numpy(), y.numpy(), z.numpy(), cmap='viridis')
ax.set_title("3D Sinc Function Surface")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z = sinc(sqrt(x² + y²))")

plt.show()
```
